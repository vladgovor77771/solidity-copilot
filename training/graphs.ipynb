```python
import glob
import json
import matplotlib.pyplot as plt
import numpy as np

results_dirs = glob.glob('./results/meta-*/segment-*/*/trainer_state.json')

# Extract the loss values from each trainer_state.json file
loss_values = []
for file_path in results_dirs:
    with open(file_path, 'r') as f:
        data = json.load(f)
        log_history = data['log_history']
        for log in log_history:
            if 'loss' in log:
                loss = log['loss']
                loss_values.append(loss)


# Calculate the interquartile range (IQR)
# q1, q3 = np.percentile(loss_values, [5, 95])
# iqr = q3 - q1

# # Calculate the lower and upper bounds for outliers
# lower_bound = q1 - (1.5 * iqr)
# upper_bound = q3 + (1.5 * iqr)

# # Remove outliers from the loss values
# loss_values = [x for x in loss_values if x >= lower_bound and x <= upper_bound]

# Plot the loss values
plt.plot(loss_values)
plt.xlabel('Iteration x100')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```
